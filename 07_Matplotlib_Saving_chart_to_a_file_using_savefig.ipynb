# Saving chart to a file using savefig

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_vals = [108,58,254,59,91]
exp_labels = ["Mayank","Pujara","Virat Kohli","Rahane","Jadeja"]
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig("piechart.jpg", bbox_inches="tight", pad_inches=1, transparent=True)
```

**We can also save in pdf format and to a different location on our computer by specifying absolute path**

```python
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig("C:/Users/chint/Downloads/py-master/py-master/matpltlib/7_save_chart/piechart.pdf", bbox_inches="tight", pad_inches=5, transparent=True)
```
