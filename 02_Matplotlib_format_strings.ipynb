# Matplotlib format strings

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y,'g+-.')
```

```python
# Same effect as 'g+-.' format string
plt.plot(x,y,color='green',marker='+',linestyle='-.')
```

**markersize**

```python
plt.plot(x,y,color='blue',marker='+',linestyle='',markersize=10)
```

**alpha property to control transparency of line chart**

```python
plt.plot(x,y,'g*',alpha=0.5) # alpha can be specified on a scale 0 to 1
```
